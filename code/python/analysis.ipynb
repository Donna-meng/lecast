```python
import csv, time
source = open("train.csv")
datareader = csv.reader(source)
datareader.next()
```

```python
interval = []
for row in datareader:
    ctstr = row[4]
    if ctstr.find('.') >= 0:
        ctstr = row[4][:ctstr.find('.')]
    ct = time.strptime(ctstr, "%Y-%m-%d %H:%M:%S")
    qtstr = row[5]
    if qtstr.find('.') >= 0:
        qtstr = row[5][:qtstr.find('.')]
    qt = time.strptime(qtstr, "%Y-%m-%d %H:%M:%S")
    itr = time.mktime(ct) - time.mktime(qt)
    interval.append(itr)
```

```python
counter = [interval.count(k) for k in range(int(min(interval)), int(max(interval))+1)]
```

```python
import matplotlib.pyplot as plt 
plt.xlabel('Count')
plt.ylabel('Second Intervals')
plt.plot(counter)
plt.show()
```
