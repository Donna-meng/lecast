```python
import csv, time
source = open("train.csv")
datareader = csv.reader(source)
datareader.next()
```

```python
clicktime = []
for row in datareader:
    ctstr = row[4]
    if ctstr.find('.') >= 0:
        ctstr = row[4][:ctstr.find('.')]
    ct = time.strptime(ctstr, "%Y-%m-%d %H:%M:%S")
    qtstr = row[5]
    if qtstr.find('.') >= 0:
        qtstr = row[5][:qtstr.find('.')]
    qt = time.strptime(qtstr, "%Y-%m-%d %H:%M:%S")
    if row[1] == '2173065':
        clicktime.append(ct.tm_hour)
```

```python
count = [clicktime.count(i) for i in range(24)]
```

```python
import matplotlib.pyplot as plt 
plt.title('Histogram of Hour Intervals of Product "2173065"')
plt.xlabel('Count')
plt.ylabel('Hour Intervals')
plt.plot(count)
plt.show()
```

```python
import csv, time
source = open("train.csv")
datareader = csv.reader(source)
datareader.next()
```

```python
def get_group(month, day):
    daycount = 0
    if month == 8:
        daycount += day - 11 + 1
    elif month == 9:
        daycount += 31 - 11 + 1 + day
    else: # month == 10
        daycount += 31-11+1 + 30 + day
    return daycount / 3 + 1
        
```

```python
clicktime = []
for row in datareader:
    ctstr = row[4]
    if ctstr.find('.') >= 0:
        ctstr = row[4][:ctstr.find('.')]
    ct = time.strptime(ctstr, "%Y-%m-%d %H:%M:%S")
    qtstr = row[5]
    if qtstr.find('.') >= 0:
        qtstr = row[5][:qtstr.find('.')]
    qt = time.strptime(qtstr, "%Y-%m-%d %H:%M:%S")
    if row[1] == '2173065':
        clicktime.append(get_group(ct.tm_mon, ct.tm_mday))
```

```python
count = [clicktime.count(i) for i in range(30)] 
plt.title('Histogram of Day Intervals of Product "2173065"')
plt.xlabel('Count')
plt.ylabel('Day Intervals')
plt.plot(count)
plt.show()
```
